```python
import xarray as xr
import os
import numpy as np
import pandas as pd
datapath = '/mnt/data/sonia/cyclone/0.5/wind_500hpa'
start_year = 1940
end_year = 1945 # inclusive!
```

```python
monthly_averages = []
for year in range(start_year, end_year+1):
    ds = xr.load_dataset(os.path.join(datapath, f'wind.{start_year}.nc'))
    correct_time = ds['time'].values[0] + pd.to_timedelta(np.arange(ds.dims['time']) * 6, unit='h')
    ds = ds.assign_coords(time=correct_time)
    monthly_average = ds.groupby('time.month').mean('time')
    monthly_averages.append(monthly_average)

monthly_stack = xr.concat(monthly_averages, dim="year")
monthly_climatology = monthly_stack.mean(dim="year")
monthly_climatology
monthly_climatology.to_netcdf(os.path.join(datapath, 'monthly_climatology.nc'))
```
